# Random function
 

```python
# for float valuew in range zero to one we have1
np.random.random()
```

```python
# for an int value in range low - high , we have
np.random.randint( low = 1, high = 20, size = [2,3,5])
```

```python
# similrily for a single row we have 
serials = np.random.randint( low = 1, high = 20, size = [20,])
serials
```

```python
# if we skip the start part , the default value will be of end
np.random.randint(5, size = [5,])
```

```python
# for creating a list of objets from a colllection of objects we have 'choice'function of randon 
list1 = ['aryan','eklavya', 'manoj','reena','shukla','alok']
```

```python
names = np.random.choice(list1, 20)
names
```

```python
marks = np.random.choice([100,90,85,80,95,75,65], 20)
```

```python
marks
```

# DataFrame

```python
# to pack the values of same indices of various iterable object like lists we can use 'zip()' function
result = list(zip(serials, names, marks ))
result
```

```python
# to represent this data in more presentable form we have 'dataframes', analyse the below given syntax
df = pd.DataFrame(data = result, columns = ('Roll number', 'Name of student', 'marks'))
df
```

```python
type(df)
```

```python
# dataframe using dictionary
df = pd.DataFrame({'Roll number' :serials, 'Name of student':names, 'Marks':marks})
```

```python
df
```

```python
# for first 5 entries of dataframe
df.head()
```

```python
# for last 5 entries
df.tail()
```

```python
# for all the values
df.values
```

```python
# for all the column names
df.columns
```

```python
# for all the values of a particular column
print(df.Marks)
# it can also be done like a ditionary element
print(df['Marks'])
```

```python
# for shape of dataframe
df.shape
```

```python
# for detailed analysis of a column data
df['Marks'].describe()     # or df.Marks.descibe()
```

```python
# for info about whole frame , use 'info()'
df.info()
```

```python
# to make a column as index we can use the 'set index command'
df.set_index('Roll number')
```

```python
df
```

```python
# as we have seen above that the roll number command is not the index value anymore, so to make it sticky we have to set the
# 'inplace' property as true as shown below
df.set_index('Roll number', inplace = True)
```

```python
df
```

```python
# to sort the dataframe according to the index value 
df.sort_index(axis = 0 , ascending = False)
```

```python
# to sort with respect to any other column we can use 'sort_value' function as shown
df.sort_values( by = 'Marks' , ascending = True)
# note that we have use 'axis = 0' which implies that the rows are to be sorted , for column operations we need to set 'axis= 1'
```

```python
# to drop a column use 
df.drop(['Marks'], axis = 1)
# 'axis = 1' as it is a column operation, and note that the index value column cant be dropped
```

```python
# the changes are not sticky or parmanent unless we use 'inplace = True'
df
```

```python
# similarily we can also drop the rows by value of index column or by indices , without even use of ' axis' parameter cuz its by default set to one
df.drop(df.index[0])
```
