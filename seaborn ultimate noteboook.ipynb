```python
import seaborn as sb
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
sb.set_style('whitegrid')
```

```python
df = pd.read_csv('global_sales_data\market_fact.csv')
```

```python
# for plotting the distribution of the data , we use distplot
# this kind of plot shows the probability distribution of a particular range of data
sb.distplot(df['Shipping_Cost'])
plt.show()
```

```python
max(list(df['Shipping_Cost']))
```

```python
# we can also plot it point by point on a rug plot which is a subset of distplot
# rug plot basically draws a line for a particular data point and it takes a lot of time to process , so u better keep the data small
sb.distplot(df['Shipping_Cost'][:200], rug = True, hist = False)
```

```python
# All the matplotlib fucntions will still work , in the backend compatability with seaborn commands
# lets plot various subplots using seaborn and matplotlib functions
plt.figure(1)

# subplot 1
plt.subplot(2,2,1)
sb.distplot(df['Sales'])
plt.show()

# subplot 2
plt.subplot(2,2,2)
sb.distplot(df['Order_Quantity'])
plt.show()

# initialise figure 2
plt.figure(2)

# subplot 3
plt.subplot(2,2,3)
sb.distplot(df['Shipping_Cost'])
plt.show()

# subplot 4
plt.subplot(2,2,4)
sb.distplot(df['Profit'])
plt.show()
```

```python
# if u want to draw grahs using two parametres then use jointplot() function as shown 
sb.jointplot(df['Sales'], df['Profit'])
plt.show()
```

```python
# to remove extreme plots we can put a condition on dataframes as
df = df[(df.Profit>-1000) & (df.Profit<1000) & (df.Sales<1500)]
sb.jointplot(df['Sales'],df['Profit'])
plt.show()
```

```python
df = df[(df.Profit>-1000) & (df.Profit<1000) & (df.Sales<1500)]
sb.jointplot(df['Sales'],df['Profit'], kind = 'hex', color = 'r')
plt.show()
```

```python
# we can plot various graphs between each pair from a list of column using pairplot() function as shown
plot = df[['Sales','Shipping_Cost','Order_Quantity','Profit']]
sb.pairplot(plot)
plt.show()
```

```python
# if we want to know the degree of corelartion berween the various columns of data then we can use ... df.corr(list of values)\
relation = plot.corr()
round(relation,3)
```

```python
# we can also plot the relation in the form of a heatmap
sb.heatmap(relation)
plt.show()
```
