```python
current  = np.linspace(1,1000, 200)
voltage = np.linspace(2,5000, 200)
```

```python
plt.plot(voltage, current)
plt.title("Ohm's Law")
plt.xlabel('current')
plt.ylabel('voltage')

# if u want to limit x,y axis values then use xlim, ylim
plt.xlim(0,3000)
plt.ylim(200,800)
plt.show()
```

```python
x = np.array(list(x for x in range(1,100,4)))
y = x.copy()
```

```python
# for changing the color and type of symbol used for creating the line we can specify the attribute  'CS' ,,, where C is color's
# first letter and S is symbol.
plt.plot(x, y**2, 'r*')
plt.show()
```

```python
# we can also draw more than one graph on a single figure example:
plt.plot(x,y,'g^'  , x, y**2, 'r*',  x, y*3, 'y+')
plt.ylim(0,1000)
```

```python
# lets plot more tha one graph in one figure
# initialising the 1st figure
plt.figure(1)

# lets partition the figure in one row and two columns for two graphs

# for subplot 1.... plt.subplot(row number, column number, graph number)
plt.subplot(1,2,1)
plt.title("Linear plot")
plt.plot(x,y)

# for subplot 2.... change plot number to 2
plt.subplot(1,2,2)
plt.title('Parabolic plot')
plt.plot(x,y**2)
```

```python
# lets create a figure with 4 subplots
plt.figure(1)

# subplot 1
plt.subplot(2,2,1)
plt.title("y = log(x)")
plt.plot(x,np.log(x))

#subplot 2
plt.subplot(2,2,2)
plt.title("y = exp(x)")
plt.plot(x, np.exp(x))

# subplot 3
plt.subplot(2,2,3)
plt.title('y = x**2')
plt.plot(x, x**2)

#subplot 4
plt.subplot(2,2,4)
plt.title('y = sqrt(64 - x**2)')
plt.plot(x,np.sqrt(10000 - x**2))
```

```python
# in this plot we see that the graphs are merging into each other , to avoid this we should initialise a new figure as shown:
# lets create a figure with 4 subplots
plt.figure(1)

plt.subplot(2,2,1)
plt.title("y = log(x)")
plt.plot(x,np.log(x))

plt.subplot(2,2,2)
plt.title("y = exp(x)")
plt.plot(x, np.exp(x))

#initialising new figure
plt.figure(2)

plt.subplot(2,2,3)
plt.title('y = x**2')
plt.plot(x, x**2)

plt.subplot(2,2,4)
plt.title('y = sqrt(64 - x**2)')
plt.plot(x,np.sqrt(10000 - x**2))

# so now we have clean and tidy graphs
```
